# Creating data

```python
movies = ['Interstellar','Inception','Infinity War','Dune','Harry Potter','Oppie','FordvsFerrari']
percentage = [9.5,8.5,9,8,9,9,7]
```

# importing Library

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Scatter Plot

```python
plt.scatter(movies, percentage)
```

# Line Plot

```python
plt.plot(movies, percentage, linestyle= 'solid')
```

```python
plt.plot(movies, percentage, linestyle= 'dashed')
```

```python
plt.plot(movies, percentage, linestyle= 'dashdot')
```

```python
plt.plot(movies, percentage, linestyle= 'dotted')
```

# Pie chart

```python
plt.pie(percentage, labels = movies)
```

# Bar Chat

```python
plt.bar(movies , percentage, linestyle= 'dotted')
```

# Sine wave

```python
import numpy as np
```

```python
time = np.arange (0,100,0.5)
time
```

```python
amplitude = np.sin(time)
amplitude
```

```python
plt.plot(time, amplitude, linestyle= 'dotted')
```
