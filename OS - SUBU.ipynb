```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

# Antecedents (inputs)
number_of_floors = ctrl.Antecedent(np.arange(1, 11, 1), 'number_of_floors')
elevator_load = ctrl.Antecedent(np.arange(0, 101, 1), 'elevator_load')
waiting_time = ctrl.Antecedent(np.arange(0, 31, 1), 'waiting_time')

# Consequent (output)
elevator_destination = ctrl.Consequent(np.arange(1, 11, 1), 'elevator_destination')

# Define fuzzy membership functions
number_of_floors.automf(3)
elevator_load['light'] = fuzz.trimf(elevator_load.universe, [0, 0, 50])
elevator_load['medium'] = fuzz.trimf(elevator_load.universe, [0, 50, 100])
elevator_load['heavy'] = fuzz.trimf(elevator_load.universe, [50, 100, 100])
waiting_time.automf(3)
elevator_destination.automf(3)

# Plot membership functions
number_of_floors.view()
elevator_load.view()
plt.show()

# Plot output membership function
elevator_destination.view()
plt.show()
```
